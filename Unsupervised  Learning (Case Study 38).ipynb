```python
# Display Image

from matplotlib import pyplot as plot
%matplotlib inline
plot.imshow(image)
```

```python
# Create Array

import numpy as np
image_np=np.asarray(image)
image_np
```

```python
# Reshape Array

image_np.shape
image_np=image_np.reshape((185*272,3))
image_np.shape
```

```python
# Elbow Method

from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++')
kmeans.fit(image_np)
wcss.append(kmeans.inertia_)
wcss    
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Sample data to demonstrate (replace with your actual data)
data = np.random.rand(100, 2)  # replace with your dataset

# Calculate WCSS for 1-10 clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

# Check if wcss has 10 values
print("Length of wcss:", len(wcss))  # Should output 10

# Plotting the elbow plot
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Sample data with 3 features for RGB-like clustering
data = np.random.rand(100, 3)  # Adjust this with your actual data

# Fit KMeans model
kmeans = KMeans(n_clusters=3, random_state=42).fit(data)

# Convert cluster centers to integers
centers = [center.astype(int) for center in kmeans.cluster_centers_]

# Create a blob
blob = np.ones((100, 100, 3))

# Ensure that each center has 3 values (for RGB)
center1_rgb = centers[0]
if center1_rgb.shape[0] == 2:
    center1_rgb = np.append(center1_rgb, [0])  # Add a third value if missing

center2_rgb = centers[1]
if center2_rgb.shape[0] == 2:
    center2_rgb = np.append(center2_rgb, [0])  # Add a third value if missing

center3_rgb = centers[2]
if center3_rgb.shape[0] == 2:
    center3_rgb = np.append(center3_rgb, [0])  # Add a third value if missing

# Create blob1, blob2, and blob3 using centers
blob1 = blob * center1_rgb
blob2 = blob * center2_rgb
blob3 = blob * center3_rgb

# Display the images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(blob1.astype(np.uint8))
plt.title("Blob1")

plt.subplot(1, 3, 2)
plt.imshow(blob2.astype(np.uint8))
plt.title("Blob2")

plt.subplot(1, 3, 3)
plt.imshow(blob3.astype(np.uint8))
plt.title("Blob3")

plt.show()
```
